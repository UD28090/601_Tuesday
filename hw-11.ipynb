```python
url = 'https://en.wikipedia.org/wiki/List_of_countries_by_past_and_projected_GDP_(nominal)'
#IMF_projections_for_2020_through_2027
```

```python
imft = pd.read_html(url, header = 0)
```

```python
len(imft)
```

```python
table2 = imft[5]
table2.head(50)
```

```python
table2.rename(columns={'Country (or dependent territory)': 'country'}, inplace=True)
table2.head(50)
```

```python
df2 = pd.merge(table2,AK,on = "country", how = 'left')
df2.head()
```

```python
df2 = df2.replace(np.nan, 0)
```

```python
df2.head(10)
```

```python
df2['average'] = ((df2['2027']-df2['2022'])*100)/df2['2022']
```

```python
df2.sort_values(by=['average'], ascending = False)
```
