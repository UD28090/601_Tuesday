# Lab

Using the flights dataframe answer the following questions

1. When do the first and last flights leave each day?

2. When there is a missing value in `dep_time` then this is an indication of canceled flight. Find the number of cancelled flights for each (`carrier`, `origin`) combination.

3. Which carrier has the worst `dep_delay`s? 

4. Which plane (`tailnum`) has the worst on-time record?

5. For each plane, count the number of flights before the first delay of greater than 1 hour.

```python
import pandas as pd
url = 'https://raw.githubusercontent.com/msaricaumbc/DS_data/master/nyc_flights.csv'

flights = pd.read_csv(url)
```

```python
#1. When do the first and last flights leave each day?

flights['month-day'] = flights['month'].astype(str)+flights['day'].astype(str)

dmin = flights.groupby(['month-day'])['dep_time'].transform(min) == flights['dep_time']
flights[dmin][['month-day','dep_time','flight']]
```

```python
dmax = flights.groupby(['month-day'])['dep_time'].transform(max) == flights['dep_time']
flights[dmax][['month-day','dep_time','flight']]
```

```python
#2.When there is a missing value in dep_time then this is an indication of canceled flight.

#Finding the number of cancelled flights for each (carrier, origin) combination.

flights[flights.dep_time.isna()][['carrier','origin']].groupby(['carrier','origin']).agg({'carrier':'count'})
```

```python
#3.Which carrier has the worst dep_delays?


f_1=flights.groupby('carrier').mean('dep_delay').sort_values(by='dep_delay')
f_1.index.tolist()[len(f_1)-1]
```

```python
#4.Which plane (tailnum) has the worst on-time record?

f_3 = flights.groupby('tailnum').agg({'dep_delay':'mean','arr_delay':'mean'})

# After calculating mean of departure and arrival delays, calculates the sum of these 2 delays to know the worst record for a tailnum

f_3['sum'] = f_3['dep_delay']+f_3['arr_delay']
f_31=f_3.sort_values(by = 'sum')
```

```python
f_32 = f_31[f_31['sum'].notna()]
f_32.index.tolist()[len(f_32)-1]
```

```python
#5. For each plane, count the number of flights before the first delay of greater than 1 hour.

df=flights.groupby('flight')['dep_delay'].apply(lambda x: (x == 60).count()).reset_index(name='count')
df
```
