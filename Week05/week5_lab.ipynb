```python
import seaborn as sns
import pandas as pd

diamonds = sns.load_dataset("diamonds")

diamonds.head()
```

#### 1.

- Plot the scatter plot of price and carat. Add clarity into the plot as hue. 

- Do you see any pattern? Is it parallel to your expectation?

```python
x = diamonds.price
y = diamonds.carat
h = diamonds.clarity 
plot = sns.scatterplot(x= x, y = y, hue = h, alpha = 0.1 )




##plot = sns.displot(data = diamonds, x= 'carat', y= 'price', hue= 'clarity', height=3 ,aspect=2)
```
